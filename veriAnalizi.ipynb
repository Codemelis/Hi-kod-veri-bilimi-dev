```python
import pandas as pd

# Veri dosyasını oku
df = pd.read_csv("survey_results_public.csv")

# Veri setinin ilk 5 satırına bakalım
df.head()
```

```python
# Türkiye'deki katılımcıları alıyoruz
turkiye_df = df[df['Country'] == 'Turkey']
print(f"Türkiye'deki katılımcı sayısı: {len(turkiye_df)}")
```

```python
# Yaş dağılımı
print(turkiye_df['Age'].value_counts().sort_index())

# Yaş dağılımını görselleştirelim
import matplotlib.pyplot as plt

age_counts = turkiye_df['Age'].value_counts().sort_index()

plt.figure(figsize=(8,5))
age_counts.plot(kind='bar', color='skyblue')
plt.title('Türkiye Yazılımcılarının Yaş Dağılımı')
plt.xlabel('Yaş Aralığı')
plt.ylabel('Katılımcı Sayısı')
plt.xticks(rotation=45)
plt.show()
```

```python
# Maaş dağılımı
print(turkiye_df['ConvertedCompYearly'].describe())

# Maaş dağılımını görselleştirelim
plt.figure(figsize=(8,5))
turkiye_df['ConvertedCompYearly'].dropna().plot(kind='hist', bins=30, color='salmon')
plt.title('Türkiye Yazılımcılarının Yıllık Gelir Dağılımı')
plt.xlabel('Yıllık Gelir (USD)')
plt.ylabel('Katılımcı Sayısı')
plt.show()
```
